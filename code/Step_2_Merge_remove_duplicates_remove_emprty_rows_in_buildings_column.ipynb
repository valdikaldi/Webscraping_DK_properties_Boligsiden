
# ---------------------------------- MERGE

```python
# path of the folder with all the csv files
input_folder = r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boligsiden_dk\Step_1_Scrape_JSON_to_table\data'

# path of the outputfile - with all the tables merged
output = r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boligsiden_dk\Step_2_Merge\Boligsiden_dk_merged_step2.csv'

# collect all the csv files in the folder of the input_folder path
csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]

# the dataframe that will collect all tables
merged_data = pd.DataFrame()

counter = 0

for csv_file in csv_files:
    counter +=1
    
    print(f'########################## file nr. {counter} out of 704 ######################')
    print(csv_file)
    print('################################################################################')
    
    #get full path
    file_path = os.path.join(input_folder, csv_file)
    
    # read in the csv file
    df = pd.read_csv(file_path, encoding='utf-8', low_memory=False)
    
    print('**data on csv file :**')
    
    # computing number of columns
    cols = len(df.axes[1])
    print(f'----------------- Number of columns csv file: {cols}')
    print()
    
    # computing number of rows
    rows = len(df.axes[0])
    print(f'----------------- Number of rows csv file: {rows}')
    
    # merge the df to the main dataframe
    merged_data = pd.concat([merged_data, df], ignore_index=True)
    
    print('**Data on MAIN dataframe: **')
    # computing number of columns
    cols = len(merged_data.axes[1])
    print(f'----------------- Number of columns csv file: {cols}')
    print()
    
    # computing number of rows
    rows = len(merged_data.axes[0])
    print(f'----------------- Number of rows csv file: {rows}')
    print()

# Save the main dataframe
merged_data.to_csv(output, encoding="utf-8", index=False)


```

# -------------- Count the number of NORMAL sales for each property 
## (With respect to the date of the sale)

```python
import ast
import datetime as dt

############### COUNT NUMBER OF NORMAL SALES IN THE 935634 PROPERTIES

print_number = 0
count_normal_sales = 0

sales_record = merged_data_no_duplicates_noBuildings_removed['registrations']

for row in sales_record:
    dictionary = ast.literal_eval(row)
    
    for sales in dictionary:
        
        if sales['type'] == 'normal':            
            # check date year
            # create date time object
            datetime = dt.datetime.strptime(sales['date'], "%Y-%m-%d") # "%d-%m-%Y")
            # convert to only date - year month and day
            date_obj = datetime.date()

            if date_obj.year > 2011:
                count_normal_sales +=1
                print_number +=1 
                
                if print_number == 15850:
                    print(count_normal_sales)
                    print_number = 0

print(f'Number of sales classified Normal and registered between 2012-2023: {count_normal_sales}')
```

# ----- GEt the number of properties with sales between 2012-2023 and normal ! 

```python
import ast
import datetime as dt

############### COUNT NUMBER OF NORMAL SALES IN THE 935634 PROPERTIES

print_number = 0
count_property = 0

sales_record = merged_data_no_duplicates_noBuildings_removed['registrations']

for row in sales_record:
    dictionary = ast.literal_eval(row)
    
    property_is_valid = False
    
    for sales in dictionary:
        
        if sales['type'] == 'normal':            
            # check date year
            # create date time object
            datetime = dt.datetime.strptime(sales['date'], "%Y-%m-%d") # "%d-%m-%Y")
            # convert to only date - year month and day
            date_obj = datetime.date()

            if date_obj.year > 2011:
                property_is_valid = True
#                 print('----its true!')
                print_number +=1
                break
                
    if property_is_valid:
        count_property +=1
        
        if print_number == 15850:
            print(count_property)
            print_number = 0

print(f'Number of sales classified Normal and registered between 2012-2023: {count_property}')
```
